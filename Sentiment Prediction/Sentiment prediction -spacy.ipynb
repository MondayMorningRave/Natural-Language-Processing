```python
!pip install spacytextblob
import spacy
from spacytextblob.spacytextblob import TextBlob

# Load the pre-trained spaCy model
nlp = spacy.load('en_core_web_sm')

# Add the SpacyTextBlob component to the pipeline
nlp.add_pipe('spacytextblob')

# Define the text to be analyzed
texts = ["The colors were bright and vivid and the house itself looks beautiful.",
         "That's about all the good things I can say about it.",
         "Let's list just some of the problems this film has!",
         "I do have to admit though--the closing line in this movie is a gem!",
         "It wasn't all that great. The book is better. Just a little lacking plot wise.",
         "A good Japanese flick for a rainy summer night."]

for text in texts:
    doc = nlp(text)
    polarity = doc._.polarity
    if polarity > 0:
        label = 'positive'
    elif polarity < 0:
        label = 'negative'
    else:
        label = 'neutral'
    print(f"The sentiment polarity of the text '{text}' is: {polarity}, and the sentiment label is: {label}")
```
