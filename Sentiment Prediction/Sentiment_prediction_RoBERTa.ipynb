```python
!pip install torch
!pip install transformers

import torch
from transformers import RobertaTokenizer, RobertaForSequenceClassification
import pandas as pd

tokenizer = RobertaTokenizer.from_pretrained('roberta-base')
model = RobertaForSequenceClassification.from_pretrained('roberta-base')
```

```python
tweets = [
    "@user @user اي علمي هاشم الي عنده بيونسيه افضل وحده jوهي اغانيها",
    "If Bernie Sanders gets elected he'll create 2nd depression by high taxes, outsource jobs,small businesses collapsing ",
    "@user @user ha! Probably google. Get some pro ones done for the next time! Off to @user tomorrow! Can't wait ",
    "十分な在庫を用意できない販売元も悪いですが、Amazon",
    "@user Por? Tenía pensado verla después de la segunda de Daredevil",
    "Se echa de menos #digital #art #life @ Escuela Superior de Arte y Tecnología (ESAT) http",
    "Rutelli: appoggio al governo #monti, sta lavorando bene #ballarò #osservatoriotivvù",
    "A Fukushima, la radioactivité met à mal la biodiversité et les oiseaux: http",
]
```

```python
results = []
for tweet in tweets:
    encoded_tweet = tokenizer.encode(tweet, padding='max_length', truncation=True, max_length=128, return_tensors='pt')
    logits = model(encoded_tweet).logits
    probabilities = torch.softmax(logits, dim=1).detach().numpy()[0]
    positive_prob = probabilities[1]
    negative_prob = probabilities[0]
    if positive_prob > negative_prob:
        sentiment = "Positive sentiment"
    else:
        sentiment = "Negative sentiment"
    results.append({'Tweet': tweet, 'Sentiment': sentiment})

df = pd.DataFrame(results)
print(df)
```
